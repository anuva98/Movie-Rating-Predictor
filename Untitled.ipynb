```python
import json
def convert_json_to_name_tag(json_str):
    tags = []
    # Parse the JSON-formatted string into a list of dictionaries
    try:
        json_obj = json.loads(json_str)
    except json.JSONDecodeError:
        print("Error decoding JSON string.")
        return []
    
    for obj in json_obj:
        name = obj["name"]
        tags.append(name)
    return tags
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD

# Example data (replace this with your own dataset)
# data = {'document_id': english_movies_df[],
#         'text': ['This is the first document.',
#                  'This document is the second document.',
#                  'And this is the third one.']}

df = english_movies_df

# Create a TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
tfidf_matrix = tfidf_vectorizer.fit_transform(df['tags'])

# pca = PCA(n_components=1000)  # Adjust n_components as needed
# tfidf_matrix_reduced = pca.fit_transform(tfidf_matrix)
n_components = 1000 
svd = TruncatedSVD(n_components=n_components)  # Adjust n_components as needed
tfidf_matrix_reduced = svd.fit_transform(tfidf_matrix)

# Get the names of the new components (columns)
component_names = [f"Component_{i+1}" for i in range(n_components)]

# Create a DataFrame with the reduced matrix and component names
df_reduced = pd.DataFrame(data=tfidf_matrix_reduced, columns=component_names)


# Combine the TF-IDF DataFrame with the original DataFrame
result_df = pd.concat([df, df_reduced], axis=1)

# Display the result

display(result_df)

```
